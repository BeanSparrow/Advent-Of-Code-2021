### Part 1

```python
class BingoBoard:
    def __init__(self, boardNumber, rowList):
        self.boardNumber = boardNumber
        self.rowList = rowList
        
    
    def checkRows(self, rowList, drawNumber):
        for row in rowList:
            for el, i in enumerate(row):
                if i == drawNumber:
                    row[el] = 'X'
    
    def checkWinner(self):
        for row in self.rowList:
            COUNT = 0
            for el, i in enumerate(row):
                if i == 'X':
                    COUNT += 1
            if COUNT == 5:
                return 1
        for row in zip(*reversed(self.rowList)):
            COUNT = 0
            for el, i in enumerate(row):
                if i == 'X':
                    COUNT += 1
            if COUNT == 5:
                return 1
        
    def winnerResult(self):
        sum = 0
        for row in self.rowList:
            for el, i in enumerate(row):
                if i != 'X':
                    sum += int(i)
        return sum
            
def parser():
    file = open('input.txt', 'r')
    drawNumbers = file.readline().split(',')
    return stripList(drawNumbers), stripList(file.read().splitlines()[1:])

def stripList(list):
    list = [line for line in list if line != '']
    list = [line.rstrip() for line in list]
    list = [line.lstrip() for line in list]
    return list

def createBoard(bingoList):
    count = 0
    boardsDict = {}
    while len(bingoList) > 0:
        rowList = [bingoList[0].split(' '), bingoList[1].split(' '), bingoList[2].split(' '), bingoList[3].split(' '), bingoList[4].split(' ')]
        for row in rowList:
            if '' in row:
                row.remove('')
        boardsDict[count] = BingoBoard(count, rowList)
        bingoList = bingoList[5:]
        count += 1
    return boardsDict


if __name__ == '__main__':
    drawNumbers, bingo = parser()
    boards = createBoard(bingo)
    winningBoard = 0
    winningNumber = 0
    for el in drawNumbers:
        for board in boards:
            boards[board].checkRows(boards[board].rowList, el)
            winBool = boards[board].checkWinner()
            if winBool:
                winningBoard = board
                winningNumber = el
                break
        if winBool:
            break
    print(boards[winningBoard].winnerResult() * int(winningNumber))
```
